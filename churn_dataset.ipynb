```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' containing the data

columns = [ 'PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
           'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
            'Churn']

# Calculate the number of rows and columns for subplots
num_rows = 7  # Change this value to adjust the number of rows
num_cols = 3  # Change this value to adjust the number of columns

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 25))
fig.subplots_adjust(hspace=0.5)  # Adjust the vertical spacing between subplots

# Iterate over the columns
for i, column in enumerate(columns):
    row = i // num_cols  # Calculate the row index
    col = i % num_cols  # Calculate the column index

    sns.countplot(data=df, x=column, ax=axes[row, col])  # Create the countplot
    axes[row, col].set_xticklabels(axes[row, col].get_xticklabels(), rotation=45)  # Rotate x-axis labels for better readability
    axes[row, col].set_title(f"Countplot of {column}")  # Set the title of the plot

# Remove empty subplots if the number of columns is not a multiple of num_cols
if i < (num_rows * num_cols) - 1:
    for j in range(i + 1, num_rows * num_cols):
        fig.delaxes(axes[j // num_cols, j % num_cols])

plt.tight_layout()  # Adjust the layout of subplots to prevent overlapping
plt.show()  # Display the plot

```
